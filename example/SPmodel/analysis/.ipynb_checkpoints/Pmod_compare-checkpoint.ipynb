```python
bonds = 4
fig, ax = plt.subplots(nrows=bonds, ncols=2, figsize=(12,4*bonds))
for bond in range(0,bonds):
    ax[bond,0].plot(PsimAC[bond,0],PsimAC[bond,1],'k',label='P sim',linewidth=1.5)
    ax[bond,0].plot(PtheAC[bond,0],PtheAC[bond,1],'b',label='P theory',linewidth=2.0)
    ax[bond,0].plot(SPspansimAC[bond,0],SPspansimAC[bond,1],'r',label='SP span',linewidth=2.0)
    ax[bond,0].set_xlim(0,100000)
    #ax[bond,0].set_ylim(0.85,1.0)
    ax[bond,0].set_xlabel('t (ps)',size=12)
    ax[bond,0].set_ylabel('$M_{1,i}(t)$',size=14)
    ax[bond,0].set_title('Bond ' + str(bond+1),size=14)
    ax[bond,0].legend()
    ax[bond,0].set_facecolor('xkcd:pale blue')
    ax[bond,0].grid(color='k', linestyle='dashed',alpha=0.15)
    ax[bond,0].set_axisbelow(True)
    #
    ax[bond,1].plot(PsimAC[bond,0],PsimAC[bond,1],'k',label='P sim',linewidth=1.4)
    ax[bond,1].plot(PtheAC[bond,0],PtheAC[bond,1],'b',label='P theory',linewidth=2.0)
    ax[bond,1].plot(SPspansimAC[bond,0],SPspansimAC[bond,1],'r',label='SP span',linewidth=2.0)
    ax[bond,1].set_xlim(1,100000)
    #ax[bond,1].set_ylim(0.85,1.0)
    ax[bond,1].set_xscale('log')
    ax[bond,1].set_xlabel('t (ps)',size=12)
    ax[bond,1].set_ylabel('$M_{1,i}(t)$',size=14)
    ax[bond,1].set_title('Bond ' + str(bond+1),size=14)
    ax[bond,1].legend()
    ax[bond,1].set_facecolor('xkcd:pale blue')
    ax[bond,1].grid(color='k', linestyle='dashed',alpha=0.15)
    ax[bond,1].set_axisbelow(True)
    
#plt.suptitle("P-SP Model Comparison of Bond Autocorrelation Functions", size=20)
plt.tight_layout()
plt.show()
plt.close()
```

```python
fig1, ax1 = plt.subplots(nrows=1, ncols=2,figsize = (12,4.5))
errb = 2
usim = PsimAC.mean(axis=0)[1]+errb*PsimAC.std(axis=0)[1]/np.sqrt(44)
lsim = PsimAC.mean(axis=0)[1]-errb*PsimAC.std(axis=0)[1]/np.sqrt(44)
uspan = SPspansimAC.mean(axis=0)[1]+errb*SPspansimAC.std(axis=0)[1]/np.sqrt(44)
lspan = SPspansimAC.mean(axis=0)[1]-errb*SPspansimAC.std(axis=0)[1]/np.sqrt(44)
uthe = PtheAC.mean(axis=0)[1]+errb*PtheAC.std(axis=0)[1]/np.sqrt(44)
lthe = PtheAC.mean(axis=0)[1]-errb*PtheAC.std(axis=0)[1]/np.sqrt(44)
ax1[0].plot(PtheAC.mean(axis=0)[0],PtheAC.mean(axis=0)[1],'b',label='P theory',linewidth=2.0)
ax1[0].plot(SPspansimAC.mean(axis=0)[0],SPspansimAC.mean(axis=0)[1],'r',label='SP span',linewidth=2.0)
ax1[0].plot(PsimAC.mean(axis=0)[0],PsimAC.mean(axis=0)[1],'k',label='P sim',linewidth=1.5)
ax1[0].fill_between(PtheAC.mean(axis=0)[0],uthe,lthe,color='b',alpha=0.25,label='theory err.')
ax1[0].fill_between(SPspansimAC.mean(axis=0)[0],uspan,lspan,color='r',alpha=0.25,label='span err.')
ax1[0].fill_between(PsimAC.mean(axis=0)[0],usim,lsim,color='g',alpha=0.25,label='sim. err.')

ax1[0].set_xlim(0,20000)
ax1[0].set_ylim(0.92,1.0)
ax1[0].set_xlabel('t (ps)',size=12)
ax1[0].set_ylabel('$M_{1,i}(t)_i$',size=14)
#ax1[0].legend(ncol=2)
ax1[0].set_facecolor('xkcd:pale blue')
ax1[0].grid(color='k', linestyle='dashed',alpha=0.18)
ax1[0].set_axisbelow(True)
#ax.set_facecolor((0.0, 0.0, 0.42))

ax1[1].plot(PtheAC.mean(axis=0)[0],PtheAC.mean(axis=0)[1],'b',label='P theory',linewidth=2.0)
ax1[1].plot(SPspansimAC.mean(axis=0)[0],SPspansimAC.mean(axis=0)[1],'r',label='SP span',linewidth=2.0)
ax1[1].plot(PsimAC.mean(axis=0)[0],PsimAC.mean(axis=0)[1],'k',label='P sim',linewidth=1.5)
ax1[1].fill_between(PtheAC.mean(axis=0)[0],uthe,lthe,color='b',alpha=0.25,label='theory err.')
ax1[1].fill_between(SPspansimAC.mean(axis=0)[0],uspan,lspan,color='r',alpha=0.25,label='span err.')
ax1[1].fill_between(PsimAC.mean(axis=0)[0],usim,lsim,color='g',alpha=0.25,label='sim. err.')
ax1[1].set_xlim(0.1,40000)
ax1[1].set_ylim(0.92,1.0)
ax1[1].set_xscale('log')
ax1[1].set_xlabel('t (ps)',size=12)
ax1[1].set_ylabel('$M_{1,i}(t)_i$',size=14)
#ax1[1].legend(ncol=2)
ax1[1].set_facecolor('xkcd:pale blue')
ax1[1].grid(color='k', linestyle='dashed',alpha=0.18)
ax1[1].set_axisbelow(True)

legend0 = ax1[0].legend(loc='upper right',ncol=2,fancybox=True,shadow=True)
frame0 = legend0.get_frame() #sets up for color, edge, and transparency
frame0.set_facecolor('xkcd:white') #color of legend
frame0.set_edgecolor('black') #edge color of legend
frame0.set_alpha(1)
legend1 = ax1[1].legend(loc='lower left',ncol=2,fancybox=True,shadow=True)
frame1 = legend1.get_frame() #sets up for color, edge, and transparency
frame1.set_facecolor('xkcd:white') #color of legend
frame1.set_edgecolor('black') #edge color of legend
frame1.set_alpha(1)
plt.suptitle('Average $M_{1,i}(t)$ over Bonds, i',size=16)
plt.tight_layout()
plt.show()
plt.close()
```
