```python
import numpy as np
import matplotlib.pyplot as plt
import os
import LE4DNA_extras
import LE4DNA
```

```python
spath = "/expanse/lustre/projects/uoo104/jeremywk/research/DNA/dsAXA/sim0/SPmodel/analysis/tcfdata/m1CAsimint_"
tpath = "/expanse/lustre/projects/uoo104/jeremywk/research/DNA/dsAXA/sim0/SPmodel/analysis/theory_tcf/m1_"
```

```python
autocorr_s = LE4DNA_extras.autocorr(92, spath)
```

```python
autocorr_t = LE4DNA_extras.autocorr(92, tpath)
```

```python
np.save('autocorr_t.npy',autocorr_t)
np.save('autocorr_s.npy',autocorr_s)
```

```python
bonds = 92
fig, ax = plt.subplots(nrows=bonds, sharey=False, ncols=2,figsize=(10,3.2*bonds))
for bond in range(0,bonds):
    #fig, ax = plt.subplots(nrows=1, ncols=1)
    ax[bond,0].plot(autocorr_s[bond][0], autocorr_s[bond][1],'k',label='SP sim',linewidth=1.0)
    ax[bond,0].plot(autocorr_t[bond][0], autocorr_t[bond][1],'r',label='SP theory',linewidth=2.0)
    ax[bond,0].set_xlim(0,1000)
    #ax[i,0].set_xscale('log')
    ax[bond,0].set_xlabel('t (ps)',size=12)
    ax[bond,0].set_ylabel('$M_{1,i}(t)$',size=14)
    ax[bond,0].set_title('Bond ' + str(bond+1),size=14)
    ax[bond,0].legend()
    ax[bond,0].set_facecolor('xkcd:pale blue')
    #ax.set_facecolor((0.0, 0.0, 0.42))
    
    ax[bond,1].plot(autocorr_s[bond][0], autocorr_s[bond][1],'k',label='SP sim',linewidth=1.0)
    ax[bond,1].plot(autocorr_t[bond][0], autocorr_t[bond][1],'g',label='SP theory',linewidth=2.0)
    ax[bond,1].set_xlim(1,40000)
    ax[bond,1].set_xscale('log')
    ax[bond,1].set_xlabel('t (ps)',size=12)
    ax[bond,1].set_ylabel('$M_{1,i}(t)$',size=14)
    ax[bond,1].set_title('Bond ' + str(bond+1),size=14)
    ax[bond,1].legend()
    ax[bond,1].set_facecolor('xkcd:pale blue')
    
plt.tight_layout()   
plt.show()
#plt.close()
```
